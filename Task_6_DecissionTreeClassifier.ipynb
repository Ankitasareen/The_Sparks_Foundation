```python
# importing required libraries in python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.datasets as datasets
```

```python
#loading the dataset
iris = datasets.load_iris()
```

```python
#forming the dataframe required
x = pd.DataFrame(iris.data , columns=iris.feature_names)
x.head(10)
```

```python
y = iris.target
y
```

## Now let us define Decision Tree Algorithm

```python
#defining the classification tree algorithm
from sklearn import tree
clftree = tree.DecisionTreeClassifier()
clftree.fit(x,y)

print("Decision tree classifier created.")
```

```python
#predicting data from the classifier
y_pred = clftree.predict(x)
y_pred
```

###### Here we can put any data of choice for prediction

```python
#creating the confusion matrix
from sklearn.metrics import accuracy_score,confusion_matrix
confusion_matrix(y,y_pred)
```

```python
accuracy_score(y,y_pred)
```

## Let us visualize the Decision Tree graphically

```python
dot_data = tree.export_graphviz(clftree , out_file = None , feature_names = x.columns , filled = True , rounded = True)
from IPython.display import Image
import pydotplus
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```
